```python
import numpy as np
def compute_cost(x,y,w,b):
  m = len(x)
  total_cost = 0
  for i in range(m):
    #prediction
    f_x = w * x[i] + b
    #error
    error = f_x - y[i]
    # squar the error
    total_cost += error**2

  cost = total_cost/(2*m) # Move calculation outside the loop
  return cost # Move return outside the loop
```

TEST THE FUNCTION

```python
x = np.array([1,2,3])
y = np.array([200,400,600])

w = 100
b = 250
cost = compute_cost(x,y,w,b)
print(cost)
```

```python
def compute_gradient(x,y,w,b):
  m = len(x)

  dj_dw = 0
  dj_db = 0

  for i in range(m):
    f_wb = w *x[i] + b
    # error
    error = f_wb - y[i]
    # gradient for w
    dj_dw += error *x [i]
    #gadient for b
    dj_db += error

  #average gradient (should be done after summing over all examples)
  dj_dw = dj_dw/m
  dj_db = dj_db/m

  return dj_dw, dj_db # Move return outside the loop
```

```python
def gradient_descent(x,y,w,b,alpha,num_iters):
  # Initialize w_hist and b_hist if you want to store history, otherwise these are not needed
  # m = len(x) is not strictly needed here as compute_gradient handles it

  for i in range(num_iters):
    #compute gradint
    dj_dw, dj_db = compute_gradient(x,y,w,b)
    #updates parameters
    w = w - alpha * dj_dw # Corrected: should use dj_dw for w update
    b = b - alpha * dj_db
    #print cost every iteration
    if i % 100 == 0:
      cost = compute_cost(x,y,w,b)
      print(f'iteration {i:4d} cost ={cost:2f}')
  return w,b # Ensure this return is executed after the loop
```

INITIALIZE PARAMETER

```python
import numpy as np

w = 0
b = 0
alpha = 0.1
iteration = 10000

# Redefine x and y to their original training values
x = np.array([1,2,3])
y = np.array([200,400,600])

#TRAIN THE MODEL
w,b = gradient_descent(x,y,w,b,alpha,iteration)
print(w,b)
```
