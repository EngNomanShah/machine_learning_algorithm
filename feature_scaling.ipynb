Method 1: Implement Min-Max Scaling from Scratch

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([
     [500,2],
     [1000,3],
     [1500,4],
     [2000,5]
    ])
```

FIND MAX AND MIN

```python
x_max = np.max(x,axis=0)
x_min = np.min(x,axis=0)
print(x_max)
print(x_min)
```

SCALE THE FEATURE

```python
x_sacled = x - x_min/x_max - x_min
print(x_sacled)
```

Method 3: Create Your Own Feature Scaling Function

```python
import numpy as np

def feature_scaling(x):
  mu =np.mean(x,axis=0)
  sigma = np.std(x,axis=0)
  x_scaled = x - mu/sigma
  return x_scaled
```

```python
x_scaled = feature_scaling(x)
print(x_scaled)
```

USE SKLEARN

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler = scaler.fit_transform(x)
print(scaler)

```

```python
plt.figure(figsize=(4,2))
plt.scatter(range(len(x)),x[:,0], label="Area")
plt.scatter(range(len(x)),x[:,1], label="Bedroom")
plt.title("Before Feature scaling")
plt.xlabel("training example")
plt.ylabel("features value ")
plt.legend()
plt.show()
```
