```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class logisticRegression():
  def __init__(self, lr=0.01, epochs=1000):
    self.lr = lr
    self.epochs = epochs
  def sigmoid(self, z):
    return  1/(1+np.exp(-z))
  def fit(self,X,y):
    m,n = X.shape
    self.w = np.zeros(n)
    self.b = 0
    cost= []
    for i in range(self.epochs):
      z = np.dot(X,self.w) + self.b
      y_hat = self.sigmoid(z)

      #gradient descent with respect to dw and db
      #dw = (1/m)  * np.dot(X.T,(y_hat-y))
      #db = (1/m) * np.sum(y_hat-y)

      #update parameter
      self.w = self.w - self.lr * (1/m)  * np.dot(X.T,(y_hat-y))
      self.b = self.b - self.lr * (1/m) * np.sum(y_hat-y)

      # loss function
      # for y = 0
      y0 = np.log(y_hat)
      #for y = 1
      y1 = np.log(1-y_hat)
      loss = 1/m * np.sum(y0+y1)
      cost.append(loss)
      plt.plot(cost, color="blue")

      # return predict probability

  def predict_prob(self,X):

    z = np.dot(X,self.w) + self.b
    return self.sigmoid(z)

  def predict(self, X):

    probability = self.predict_prob(X)
    return (probability >= 0.5).astype(int)


```

TRAINING THE DATA

```python
import numpy as np
#training data
X = np.array([[1,2],
             [2,3],
             [3,4],
             [4,5]])

#labels
y = np.array([0,0,1,1])

#creat a model
model = logisticRegression(lr=0.1,epochs=1000)


#training model
model.fit(X,y)

prediction = model.predict(X)

probability = model.predict_prob(X)
print("prediction", prediction)
print("probability", probability)
```
